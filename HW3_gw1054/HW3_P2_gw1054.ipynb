```python
import pandas as pd

#Read csv from opendata nyc

df = pd.read_csv('https://data.cityofnewyork.us/resource/4qxi-jgbe.csv')

#display the dataframe table

df.head(7)
```

```python
#Select two columns of numerica values

df_num = df[['regular_gross_paid', 'regular_hours']].copy()
df_num = df_num.dropna()
```

```python
#Display the selected columns

df_num.head()
```

```python
import matplotlib.pyplot as plt

#plot the numerica data with pyplot

%matplotlib inline

df_num.plot(['regular_hours'], ['regular_gross_paid'], fontsize=10, kind='scatter', figsize=(12,10),
           title='Regular gross paid VS. Regular working hours', legend=True)
```

## Caption:
The regular gross pay icreases as the regular working hours increase; but at the margin of around 1700 hours, different agencies are paid at drastically different rates.

```python
#Extra Credit

ec = pd.read_csv('https://data.cityofnewyork.us/resource/uvbq-3m68.csv')
ec.head(7)
```

```python
ec_num = ec[['fine_amount', 'issue_date']].copy()
ec_num = ec_num.dropna()
ec_num.head()
```

```python
%matplotlib inline

ec_num['issue_date'] = pd.to_datetime(ec_num['issue_date'])
ec_num.plot(['issue_date'], ['fine_amount'], fontsize=10, kind='scatter', figsize=(12,10),
            title='Fine amount VS. Issued date', legend=True, rot=90)
```

## Caption:
The parking violation data is missing from around 2018-09 to 2017-03

Parking fines are usually below $120, except for one case in 2014
