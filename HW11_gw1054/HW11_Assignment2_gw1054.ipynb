```python
import urllib
import pandas as pd
import os
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
url = 'https://data.cityofnewyork.us/resource/waf7-5gvc.json'
data = pd.read_json(url)
```

```python
data = data.sort_values('year')
```

```python
data.head()
```

```python
data.rename(columns={'new_york_city_population': 'population', 'nyc_consumption_million_gallons_per_day': 'nyc_consumption',
                    'per_capita_gallons_per_person_per_day': 'consumption_per_person'}, inplace=True)
```

```python
data.population /= 1000000
data.nyc_consumption /= 100
```

```python
fig, ax = plt.subplots(figsize=(10,10))
# ax2 = ax.twinx()
# x = data.year
# y1 = data.population
# y2 = data.consumption_per_person
# ax.plot(x, y1, 'g-')
# ax2.plot(x, y2, 'b-')

p1 = data.plot(x='year', y='population', ax=ax)
p2 = data.plot(x='year', y='nyc_consumption', ax=ax, secondary_y=True, ylim=(7,16))
plt.title("NYC Water Consumption and Population From 1979 to 2017", fontsize=20)
ax.set_ylabel("Population (Million)", fontsize=16)
plt.ylabel("Water Consumption (100 Million Gallons per Day)", fontsize=16)
ax.set_xlabel("Year", fontsize=16)
```
